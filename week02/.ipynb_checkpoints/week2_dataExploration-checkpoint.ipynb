```python
areas_highest_need = parks.query('NEED_DESCP == "Very High"')
```

```python
print(areas_highest_need[['STUD_AR_NM', 'NEED_DESCP']])
```

## but here I ran a query to find a list of the areas with 'Very High' park need

```python
areas_highest_need_district = parks.query('NEED_DESCP == "Very High"')[['STUD_AR_NM', 'DISTRICT']]
```

```python
print(areas_highest_need_district)
```

```python
# created a column to see which Supervisorial District each area resides in
```

```python
very_high_need_district_totals = parks.query('NEED_DESCP == "Very High"').groupby('DISTRICT').size()
```

```python
print(very_high_need_district_totals)
```

```python
# here is the amount of high need areas per Supervisorial District
```

```python
very_high_need_district_totals.plot()
```

```python
# plotting the district total
```

```python
very_high_need_district_totals.plot(kind='bar', color='green', title = 'Total Very High Need Areas per District')
```
